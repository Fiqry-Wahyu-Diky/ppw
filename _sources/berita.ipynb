# **Crawling berita**

## 1.Install package dan requirements

```python
# requirements
!pip install requests
import requests
from bs4 import BeautifulSoup
import csv
```

## 2.Set url yang dibutuhkan

```python
url = "https://www.detik.com/"
```

## 3.Set kategori yang dicrawling

```python
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser') #Isi teks dari respons HTTP yang diterima dari server web setelah melakukan permintaan GET.
pages = soup.findAll('article', {'class' :'list-content__item'})
first_page = soup.findAll('li',"nav__item")


list_a = []
for items in first_page:
  # ===== mengambil beberapa navbar buat diakses =====
  cek_a = items.findAll("a")
  if cek_a:
    a = items.find("a").get('href')
    list_a.append(a)

print(list_a)
# ==== mengambil 3 indeks berita ========
listberita = [list_a[4],list_a[7],list_a[8]]
```

## 4.Proses Crawling

```python
datas = []

# ==== perulangan pages berita =====
for ipages in range(1,16):
  # ==== mengambil link setiap list berita ===
  for beritas in listberita:
    response_berita = requests.get(beritas+"indeks/"+str(ipages))

    namecategori = beritas.split("//")[1].split(".")
    print(f"... Mengambil data berita detik {namecategori[0]}, Pages ke-{ipages}")

    soup_berita = BeautifulSoup(response_berita.text, 'html.parser')
    pages_berita = soup_berita.findAll('article', {'class' :'list-content__item'})

    for items in pages_berita:
      cek_link_in = items.findAll("h3","media__title")
      if cek_link_in:
        link_in = items.find("h3","media__title")
      get_link_in = link_in.find("a").get("href")
      # print(get_link_in)
      # ======== masuk ke link pages ==========
      response_artikel = requests.get(get_link_in)
      soup_artikel = BeautifulSoup(response_artikel.text, 'html.parser')

      # =========== judul =========
      cek_judul = soup_artikel.findAll("h1","detail__title")
      if cek_judul:
        judul = soup_artikel.find("h1","detail__title").text.replace("\n","")
      else:
        judul=""
      # print("judul : ",judul)

      # ========== label ==========
      # cek_label = soup_artikel.find_all("span","detail__label")
      # # print(cek_label)
      # if cek_label:
      #   label = soup_artikel.find("span","detail__label").text.replace("Sepakbola","Sport").replace("\n","")
      # else:
      #   label = None
      label = namecategori[0]

      # ======= tempat ==========
      cek_location = soup_artikel.find_all("strong")
      if cek_location:
        location = soup_artikel.find("strong").text.replace("\n","")
      else:
        location = ""

      # =========== author =========
      cek_author = soup_artikel.find_all("div","detail__author")
      # print(cek_author)
      if cek_author:
        author = soup_artikel.find("div","detail__author").text.replace(label,"").replace(" -",'').replace("\n","")
      else:
        author = None
      # print(author)

      # ========== date =============
      cek_date = soup_artikel.findAll("div","detail__date")
      if cek_date:
        date = soup_artikel.find("div","detail__date").text.replace("\n","")
      else:
        date = ""
      # print(date)

      # ======= cek iklan ======
      cek_iklan = soup_artikel.find_all("div","parallaxindetail scrollpage")
      if cek_iklan:
        iklan = soup_artikel.find("div","parallaxindetail scrollpage").text.replace("\n","")


      # ====== cek lainnya ======
      cek_link_sisip = soup_artikel.find_all("table","linksisip")
      if cek_link_sisip:
        link_sisip = soup_artikel.find("table","linksisip").text.replace("\n","")
      else:
        link_sisip = ""

      cek_bodytag = soup_artikel.find_all("div","detail__body-tag mgt-16")
      if cek_bodytag:
        bodytag = soup_artikel.find("div","detail__body-tag mgt-16").text.replace("\n","")
      else:
        bodytag = ""

      cek_advertise = soup_artikel.find_all("div","ads-scrollpage-height")
      if cek_advertise:
        advertise = soup_artikel.find("div","ads-scrollpage-height").text
        # print(advertise)

      # ========== artikel ===========
      cek_artikel = soup_artikel.find_all("div","detail__body-text itp_bodycontent")
      if cek_artikel:
        arrArtikel = soup_artikel.findAll("p")
        artikel = ""
        for artikels in arrArtikel:
          text = artikels.get_text().replace(iklan,"").replace(link_sisip,"").replace(bodytag,"").replace(location,"").replace("\n","").replace(" -",'').replace("\r\t\t\t\t\t\t\t","").replace(advertise,"").replace("\n\rADVERTISEMENT\r\n\rSCROLL TO CONTINUE WITH CONTENT\r\n","")
          artikel += text + "\n"
      else:
        artikel = None
      # print(artikel)

      datas.append({
        'Tanggal' : date,
        'Penulis':author,
        "Tempat" : location,
        'Judul': judul,
        'Artikel': artikel,
        'Label' :  label
            })
```

```python
datas
```

## 5.Simpan data dalam format CSV

```python
# Menyimpan data dalam bentuk CSV
csv_filename = 'data_crawling_berita_small.csv'
with open(csv_filename, 'w', newline='', encoding='utf-8') as csv_file:
    fieldnames = ['Tanggal', 'Penulis','Tempat', 'Judul', 'Artikel', 'Label']
    csv_writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    csv_writer.writeheader()

    for entry in datas:
        csv_writer.writerow(entry)

print(f"Data telah disimpan dalam file {csv_filename}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Merge lists of strings into a single string per element in the column
dataBerita['final_artikel'] = dataBerita['tokenizing'].apply(lambda x: ' '.join(x))

# Initialize the TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Transform the column using the TF-IDF Vectorizer
tfidf_matrix = tfidf_vectorizer.fit_transform(dataBerita['final_artikel'])

# Convert the TF-IDF matrix to a DataFrame
feature_names = tfidf_vectorizer.get_feature_names_out()
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)

# Display the DataFrame
tfidf_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Inisialisasi DataFrame untuk Term Frequency (TF)
# df = pd.DataFrame(data)

# Inisialisasi CountVectorizer
vectorizer = CountVectorizer()

# Melakukan transformasi TF pada kolom 'final_abstrak'
tf_matrix = vectorizer.fit_transform(dataBerita['final_artikel'])

# Membuat DataFrame dari hasil TF
tf_df = pd.DataFrame(tf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

# Cetak DataFrame TF
tf_df
```
