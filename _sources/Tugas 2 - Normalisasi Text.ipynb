### **Punctuation process**

> **Punctuation process** merupakan proses normalisasi data yang bertujuan untuk menghilangkan tanda baca



```python
# recruitments punctuation
import string
```

```python
# membuat kolom baru dengan nama new_abstrak untuk data baru yang dipunctuation

data['new_abstrak'] = data['Abstrak'].str.replace('[{}]'.format(string.punctuation), '').str.lower()
```

```python
# string.punctuation
```

```python
#========
# Menghilangkan angka dari kolom 'new_abstrak'
data['new_abstrak'] = data['new_abstrak'].str.replace('\d+', '', regex=True)
```

### **Stopword**

> Stopwords digunakan untuk menghilangkan kata umum yang sering muncul dalam teks seperti: di, dan, atau, dari, ke, saya.



```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
import nltk
nltk.download('punkt')

# Download kamus stop words
nltk.download('stopwords')
```

```python
# Inisialisasi kamus stop words dari NLTK
stop_words = set(stopwords.words('indonesian'))  # Inisialisasi kamus stop words

# Menghapus stop words dari kolom 'Abstrak'
for stop_word in stop_words:
  data['new_abstrak_stopword'] = data['new_abstrak'].str.replace(rf'\b{stop_word}\b', '', regex=True) #rf untuk formating string
```

### Tokenizing



> **Tokenizing** adalah proses memecah teks atau dokumen menjadi potongan-potongan yang lebih kecil, yang disebut token.



```python
data['tokens'] = data['new_abstrak'].apply(word_tokenize)
```

### stemming



> merupakan proses normalisasi data teks menjadikan kata dasar



```python
!pip install sastrawi
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Membuat kolom baru untuk hasil stemming
data['stemmed_tokens'] = None

# Melakukan stemming pada setiap elemen dalam kolom 'tokens' tanpa fungsi atau list comprehension
for index, row in data.iterrows(): #perulangan indeks, row dari data
    stemmed_tokens = []
    for word in row['tokens']:
      # print(word)
      stemmed_word = stemmer.stem(word)
      # print(stemmed_word)
      stemmed_tokens.append(stemmed_word)
    # print(stemmed_tokens)
    data.at[index, 'stemmed_tokens'] = stemmed_tokens



# data
```

### Binary

> Binary merupakan sebuah pengolahan data yang bertujuan agar dapat digunakan oleh algoritma machine learning. Binary akan mengkonversi kata unik menjadi vektor binner yaitu 1 0.



```python
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.preprocessing import OneHotEncoder
```

```python
# Membuat DataFrame kosong dengan kolom sesuai kata unik
teks_uniqe = set()
for word in data['stemmed_tokens']:
  teks_uniqe.update(word)

teks_uniqe_list = list(teks_uniqe) #menjadikan ke list
```

```python
teks_uniqe_list
```

```python
df_baru = pd.DataFrame(columns=teks_uniqe_list)

# Mengisi DataFrame baru dengan data['stemmed_tokens'] secara manual
for tokens in data['stemmed_tokens']:
    row_data = []
    for token in teks_uniqe_list:
        if token in tokens:
            row_data.append(1)
        else:
            row_data.append(0)
    df_baru = df_baru.append(pd.Series(row_data, index=teks_uniqe_list), ignore_index=True)
```

```python
# df_baru
hasil_encode = pd.concat([pd.DataFrame(data['Abstrak'], columns=['Abstrak']), df_baru], axis=1)
hasil_encode
```

### Term Freq

> Term Freq adalah konsep yang digunakan dalam pemrosesan teks dan analisis teks untuk mengukur sejauh mana suatu kata atau term muncul dalam sebuah dokumen atau koleksi dokumen. Term frequency menggambarkan seberapa sering sebuah kata muncul dalam teks relatif terhadap total kata dalam dokumen tersebut.






```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
data['stemmed_tokens']
```

```python
# Inisialisasi DataFrame untuk Term Frequency (TF)
df_tf = pd.DataFrame(columns=teks_uniqe_list)

# Menghitung Term Frequency (TF) secara manual untuk setiap dokumen
for word_list in data['stemmed_tokens']:
  tf_data = []
  for word in teks_uniqe:
      tf_count = 0
      for word_in_doc in word_list:
          if word == word_in_doc:
              tf_count += 1
      tf_data.append(tf_count)
  df_tf = df_tf.append(pd.Series(tf_data, index=teks_uniqe), ignore_index=True)

```

```python
df_tf.isna().sum()
```

```python
hasil_tf_df = pd.concat([pd.DataFrame(data['stemmed_tokens'], columns=['stemmed_tokens']), df_tf], axis=1)
hasil_tf_df
```

```python
hasil_tf_df.to_csv('tf_df.csv', index=False)
```

### Log Freq

>  Log Freq adalah perhitungan bobot kata atau istilah (term weights) dalam representasi vektor dokumen. Ini bertujuan untuk mengurangi dampak kata-kata yang muncul sangat sering dalam koleksi dokumen, sehingga kata-kata tersebut tidak mendominasi bobot vektor.



```python
import numpy as np
```

```python
df_tf
```

```python
# Mengonversi tipe data kolom menjadi float
df_tf = df_tf.astype(float)

# Menghitung Logarithmic Term Frequency (Log-TF)
df_log_tf = np.log(df_tf + 1)  # Menggunakan np.log dengan penambahan 1
```

```python
hasil_df_log_tf = pd.concat([pd.DataFrame(data['Abstrak'], columns=['Abstrak']), df_log_tf], axis=1)
hasil_df_log_tf
```

### TF-IDF



> metode yang digunakan dalam pemrosesan bahasa alami (Natural Language Processing - NLP) dan pengambilan informasi (Information Retrieval) untuk mengukur seberapa penting suatu kata dalam suatu dokumen atau kumpulan dokumen. Tujuannya adalah untuk memberikan bobot yang lebih tinggi kepada kata-kata yang penting dalam representasi teks.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Menggabungkan kata-kata dalam daftar menjadi satu string
# data['stemmed_tokens'] = data['stemmed_tokens'].apply(lambda tokens: ' '.join(tokens))
data_teks = []
for i in data['stemmed_tokens']:
  data_i = " ".join(i)
  data_teks.append(data_i)
# Menginisialisasi objek TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Menghitung TF-IDF dari kolom "stemmed tokens"
tfidf_matrix = tfidf_vectorizer.fit_transform(data_teks)
```

```python

# Anda dapat mengonversi hasilnya menjadi DataFrame jika diperlukan
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
hasil_df_tfidf = pd.concat([pd.DataFrame(data['Abstrak'], columns=['Abstrak']), tfidf_df], axis=1)
hasil_df_tfidf
```

```python
# Menyimpan DataFrame ke dalam berkas CSV
hasil_df_tfidf.to_csv('df_tfidf.csv', index=False)
```
